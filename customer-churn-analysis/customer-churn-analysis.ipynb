```python
# Convert TotalCharge from string to numeric

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
# Remove rows with NaN

df2= df.dropna()
```

```python
df2 = df2.drop("customerID", axis = 1)
```

```python
# select features and targets

x = df2.drop("Churn", axis = 1)
y = df2["Churn"]
```

```python
# Encode categorical variable

x_encoded = pd.get_dummies( x, drop_first = True )
```

```python
# Importing the necessary tools 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
x_train, x_test, y_train, y_test = train_test_split( 
    x_encoded, y, test_size = 0.2, random_state = 42
)
```

```python
# Rescaling for standardization
# The model initially failed to converge due to unscaled features with different magnitudes. 
# Applying standardization significantly improved convergence and stability, allowing logistic regression to train efficiently.

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
model = LogisticRegression(max_iter = 1000)

model.fit(x_train_scaled, y_train)
```

```python
model.n_iter_
```

```python
y_pred = model.predict(x_test_scaled)
y_pred
```

```python
#Accuracy check

accuracy = accuracy_score(y_pred, y_test)
accuracy
```

```python
# Confusion matrix to check how often is the model correct

cm = confusion_matrix( y_pred, y_test)
cm
```

```python
print(classification_report(y_pred, y_test))
```

```python
# Feature_importance

feature_importance = pd.Series(
    model.coef_[0],
    index=x_encoded.columns
).sort_values()

feature_importance.head(20)
```
