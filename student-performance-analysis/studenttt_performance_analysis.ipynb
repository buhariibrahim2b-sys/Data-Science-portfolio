```python
# Checking the impact of parental level of education on average score

df.groupby("parental level of education")[["average score"]].mean()
```

```python
# Bar chart showiing the impact of parental level of education on average score

plt.figure(figsize=(8, 4))
sns.barplot(
    data=df,
    x="parental level of education",
    y="average score",
    estimator="mean"
)
plt.xticks(rotation=45)
plt.title("Average Score by Parental Education")
plt.show()
```

```python
# Importing tools

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Selecting feature(x) and target(y)

x = df[["reading score","writing score"]]
y= df["math score"]
```

```python
# Train and test split

x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.2, random_state = 42)
```

```python
# Check trian size

x_train.shape
```

```python
y_train.shape
```

```python
# Train model

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Model parameters (coefficients)

model.coef_
```

```python
# Model parameter ( intercept)

model.intercept_
```

```python
# Model prediction

y_pred = model.predict(x_test)
y_pred
```

```python
# Model evaluation ( mean squsre error)

mean_squared_error(y_pred, y_test)
```

```python
# Model evaluation ( r2)

r2_score(y_pred, y_test)
```
